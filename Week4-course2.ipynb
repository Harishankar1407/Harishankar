```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook
```

```python
bahrain = pd.read_csv(r'C:\Users\Asus\Downloads\Bahrain\Bahrain.csv')
bahrain.drop(bahrain.columns[3], inplace=True, axis=1)
bahrain=bahrain.set_index('Year')
bahrain.head()
```

```python
india = pd.read_csv(r'C:\Users\Asus\Downloads\India\India.csv')
india.drop(india.columns[3], inplace=True, axis=1)
india=india.set_index('Year')
india.head()
```

```python
df = pd.merge(bahrain, india, on='Year', how='outer')
df.head()
```

```python
df = df.rename(columns={"Country or Area_x": "Bahrain", "Value_x": "Value in bahrain", "Country or Area_y": "India",
                   "Value_y": "Value in India"})
df=df.sort_index(axis=0, level=None, ascending=True)
df.head()
```

```python
plt.figure()
df.plot(colormap='Set1');
plt.xlabel('Years')
plt.ylabel('Population')
plt.title('Percentage of individuals using the Internet in India and Bahrain')
plt.show()
```
